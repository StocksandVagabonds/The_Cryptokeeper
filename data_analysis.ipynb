Now that we've thoroughly integrated our datasets and prepped the data, we can start running some models.\
Since we want the predictive outcome to be continuous rather than categorical, we will use a linear regression rather than a logistic regression model.

```python
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split

#import python script that I made which contains helper functions
import pipeline as pl
```

First model will look at the entire dataset and see how elon tweeting, and google searches relating to elon and doge can be used to predict prices.\
Consider normalizing the features as well.

```python
dogdf.columns
```

```python
dogdf = dogdf[dogdf['Month'] != '2021-10']
dogdf = dogdf[dogdf['Month'] != '2021-09']
dogdf = dogdf[dogdf['Month'] != '2021-08']
dogdf = dogdf[dogdf['Month'] != '2021-07']
dogdf = dogdf[dogdf['Month'] != '2021-06']
dogdf = dogdf[dogdf['Month'] != '2021-05']
dogdf = dogdf[dogdf['Month'] != '2021-04']
```

```python
train, test = train_test_split(dogdf[['elon doge: (United States)', 'elon_tweet', 'percent_change']], test_size = 0.3, random_state = 25)
```

```python
pl.fill_all_missing(train)
```

```python
pl.fill_all_missing(test)
```

```python
train, test = pl.normalize(train, test, ['elon doge: (United States)', 'elon_tweet'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics

lin_reg = linear_model.LinearRegression()

x = train[['elon doge: (United States)', 'elon_tweet']]
lin_reg.fit(x, train['percent_change'])

best_line_multi = lin_reg.predict(test[['elon doge: (United States)', 'elon_tweet']])
features = ['elon doge: (United States)', 'elon_tweet']

plt.scatter(test['percent_change'], best_line_multi, color='blue', s=10)
plt.title('Linear Regression, Daily Price Action (01/2019 - 03/2021)')
plt.xlabel('Test values')
plt.ylabel('Predicted Values')
plt.show()

print('The MAE is', metrics.mean_absolute_error(test['percent_change'], best_line_multi))
print('The MSE is', metrics.mean_squared_error(test['percent_change'], best_line_multi))
print('The R^2 value is', metrics.r2_score(test['percent_change'], best_line_multi))

importance = lin_reg.coef_
for i,v in enumerate(importance):
    print('Feature: {}, Score: %.5f'.format(features[i]) % (v))
```
